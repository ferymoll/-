## Лабораторная работа 3.5

Вычислить определенный интеграл, методами прямоугольников, трапеций, Симпсона с шагами h1, h2. Оценить погрешность вычислений, используя  Метод Рунге-Ромберга  

```python
import numpy as np


def f(x):
    return 1 / (3 * x ** 2 + 4 * x + 2)


def rectangle_rule_integration(f, a, b, h):
    xk = np.arange(a, b + h, h)
    y = np.empty(len(xk)-1)
    for i in range(len(xk)-1):
        y[i] = f((xk[i] + xk[i+1]) / 2)
    F = np.sum(y) * h
    return F


def trapezoidal_rule_integration(f, a, b, h):
    xk = np.arange(a, b + h, h)
    y = np.empty(len(xk))
    for i in range(len(xk)):
        y[i] = f(xk[i])
    F = (np.sum(y) - y[0]/2 - y[-1]/2) * h
    return F


def simpsons_rule_integration(f, a, b, h):
    xk = np.arange(a, b + h, h)
    y = np.empty(len(xk))
    for i in range(len(xk)):
        y[i] = f(xk[i])
    F = (y[0] + y[-1] + 4 * np.sum(y[1:-1:2]) + 2 * np.sum(y[2:-1:2])) * h/3
    return F


def runge_romberg_error(integral_h1, integral_h2, p):
    return np.abs(integral_h1 - integral_h2) / (2**p - 1)


x0 = -2
xk = 2
h1 = 1.0
h2 = 0.5


integral_rect_h1 = rectangle_rule_integration(f, x0, xk, h1)
integral_trap_h1 = trapezoidal_rule_integration(f, x0, xk, h1)
integral_simp_h1 = simpsons_rule_integration(f, x0, xk, h1)


integral_rect_h2 = rectangle_rule_integration(f, x0, xk, h2)
integral_trap_h2 = trapezoidal_rule_integration(f, x0, xk, h2)
integral_simp_h2 = simpsons_rule_integration(f, x0, xk, h2)


error_rect = runge_romberg_error(integral_rect_h1, integral_rect_h2, p=2)
error_trap = runge_romberg_error(integral_trap_h1, integral_trap_h2, p=2)
error_simp = runge_romberg_error(integral_simp_h1, integral_simp_h2, p=4)

def runge_romberg(integral_h1, integral_h2, p):
    return integral_h2 + np.abs(integral_h1 - integral_h2) / (2**p - 1)


print("Метод прямоугольников:")
print("Значение интеграла на отрезке с шагом h = {0}: {1}".format(h1, integral_rect_h1))
print("Значение интеграла на отрезке с шагом h = {0}: {1}".format(h2, integral_rect_h2))
print("Уточнение значения методом Рунге-Ромберга: {0}".format(runge_romberg(integral_rect_h1, integral_rect_h2, 2)))
print("Погрешность: {0}\n".format(abs(error_rect)))
print()

print("Метод трапеций:")
print("Значение интеграла на отрезке с шагом h = {0}: {1}".format(h1, integral_trap_h1))
print("Значение интеграла на отрезке с шагом h = {0}: {1}".format(h2, integral_trap_h2))
print("Уточнение значения методом Рунге-Ромберга: {0}".format(runge_romberg(integral_trap_h1, integral_trap_h2, 2)))
print("Погрешность: {0}\n".format(abs(error_trap)))
print()


print("Метод Симпсона:")
print("Значение интеграла на отрезке с шагом h = {0}: {1}".format(h1, integral_simp_h1))
print("Значение интеграла на отрезке с шагом h = {0}: {1}".format(h2, integral_simp_h2))
print("Уточнение значения методом Рунге-Ромберга: {0}".format(runge_romberg(integral_simp_h1, integral_simp_h2, 2)))
print("Погрешность: {0}\n".format(abs(error_simp)))
```
