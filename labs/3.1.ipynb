## Лабораторная работа 3.1 

Используя таблицу значений Yi функции y = f(x), вычисленных в точках
Xi, i = 0,...,3 построить интерполяционные многочлены Лагранжа и Ньютона, проходящие через точки {Xi, Yi}. Вычислить значение погрешности интерполяции в точке X*

```python
import sympy as sp

x = sp.Symbol('x')


def Lagrange(f, X, x_st):
    n = len(X)
    Y = []
    for i in range(n):
        Y.append(f.evalf(subs={x: X[i]}))
    Y_st = f.evalf(subs={x: x_st})

    omega_x = sp.S.One
    Omega = []
    for i in range(n):
        omega_x *= x - X[i]
        om = 1
        for j in range(n):
            if i != j:
                om *= X[i] - X[j]
        Omega.append(om)
    sp.simplify(omega_x)

    L_str = ""
    L_x = sp.S.Zero
    for i in range(n):
        f_O = Y[i] / Omega[i]
        if sp.sign(f_O) == 1:
            L_str += f"+{omega_x / (x - X[i]) * f_O.evalf(5)}"
        else:
            L_str += f"{omega_x / (x - X[i]) * f_O.evalf(5)}"
        L_x += omega_x / (x - X[i]) * f_O

    L_st = L_x.evalf(subs={x: x_st})
    d = abs(Y_st - L_st)
    a = L_st

    return L_str, d, a


def Newton(f, X, x_st):
    n = len(X)
    F = []
    for i in range(n):
        F.append(f.evalf(subs={x: X[i]}))

    F1 = []
    for i in range(n-1):
        F1.append(((F[i] - F[i+1]) / (X[i] - X[i+1])).evalf(5))
    # print("F1: {0}".format(F1))
    F2 = []
    for i in range(n-2):
        F2.append((F1[i] - F1[i+1]) / (X[i] - X[i+2]))
    # print("F2: {0}".format(F2))
    F3 = []
    for i in range(n-3):
        F3.append((F2[i] - F2[i+1]) / (X[i] - X[i+3]))
    # print("F3: {0}".format(F3))

    P_x = F[0] + (x - X[0]) * F1[0] + (x - X[0]) * (x - X[1]) * F2[0]\
        + (x - X[0]) * (x - X[1]) * (x - X[2]) * F3[0]

    P_str = ""
    P_str += f"{F[0].evalf(5)}"
    if sp.sign(F1[0].evalf(5)) == 1:
        P_str += f" +{F1[0].evalf(5)} * ({x - X[0]})"
    else:
        P_str += f" {F1[0].evalf(5)} * ({(x - X[0])})"

    if sp.sign(F2[0].evalf(5)) == 1:
        P_str += f" +{F2[0].evalf(5)} * {(x - X[0]) * (x - X[1])}"
    else:
        P_str += f" {F2[0].evalf(5)} * {(x - X[0]) * (x - X[1])}"
    if sp.sign(F3[0].evalf(5)) == 1:
        P_str += f" +{F3[0].evalf(5)} * {(x - X[0]) * (x - X[1]) * (x - X[2])}"
    else:
        P_str += f" {F3[0].evalf(5)} * {(x - X[0]) * (x - X[1]) * (x - X[2])}"

    P_st = P_x.evalf(subs={x: x_st})
    F_st = f.evalf(subs={x: x_st})
    d = abs(F_st - P_st)
    a = P_st

    return P_str, d, a


y = sp.sqrt(x)
X_a = [0, 1.7, 3.4, 5.1]
X_b = [0, 1.7, 4.0, 5.1]
X_star = 3.0

# ЛАГРАНЖ
L_a, delta_L_a, a = Lagrange(y, X_a, X_star)
print("Интерполяционный многочлен Лагранжа по набору А:\nL(x) = {0}".format(L_a))
print("Погрешность интерполяции(|y(3) - L(3)|): {0}".format(delta_L_a))
print("Y(X*) = {0}\n".format(a))

L_b, delta_L_b, b = Lagrange(y, X_b, X_star)
print("Интерполяционный многочлен Лагранжа по набору B:\nL(x) = {0}".format(L_b))
print("Погрешность интерполяции(|y(3) - L(3)|): {0}".format(delta_L_b))
print("Y(X*) = {0}\n".format(b))

print('-------------------------------------------------------------------------------\n')

# НЬЮТОН
P_a, delta_P_a, N_a = Newton(y, X_a, X_star)
print("Интерполяционный многочлен Ньютона по набору А:\nP(x) = {0}".format(P_a))
print("Погрешность интерполяции(|y(3) - P(3)|): {0}".format(delta_P_a))
print("Y(X*) = {0}\n".format(N_a))

P_b, delta_P_b, N_b = Newton(y, X_b, X_star)
print("Интерполяционный многочлен Ньютона по набору B:\nP(x) = {0}".format(P_b))
print("Погрешность интерполяции(|y(3) - P(3)|): {0}".format(delta_P_b))
print("Y(X*) = {0}\n".format(N_b))
```
